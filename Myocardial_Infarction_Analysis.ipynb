```python
import pandas as pd
import numpy as np

# Function to impute missing values
def impute_columns(data):
    # Convert all columns to numeric if possible
    for col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')

    # Define imputation strategies for specific columns
    columns_mode = [
        "ZSN_A", "endocr_01", "NA_R_3_n", "zab_leg_01", "n_r_ecg_p_06",
        "ritm_ecg_p_02", "R_AB_3_n", "NOT_NA_1_n", "np_09", "NA_R_1_n", "ANT_CA_S_n", "GB"
    ]
    columns_mean = ["AGE", "lat_im", "S_AD_ORIT", "AST_BLOOD"]
    columns_median = ["nr_04", "MP_TP_POST", "nr_01", "ALT_BLOOD", "ROE"]

    # Apply mode imputation
    for col in columns_mode:
        if col in data.columns:
            data[col].fillna(data[col].mode()[0], inplace=True)

    # Apply mean imputation
    for col in columns_mean:
        if col in data.columns:
            data[col].fillna(data[col].mean(), inplace=True)

    # Apply median imputation
    for col in columns_median:
        if col in data.columns:
            data[col].fillna(data[col].median(), inplace=True)

    # Impute remaining missing values with column-specific strategies if necessary
    additional_columns = {
        "DLIT_AG": data["DLIT_AG"].median(),
        "nr_07": data["nr_07"].mode()[0],
        "NOT_NA_KB": data["NOT_NA_KB"].median(),
        "TRENT_S_n": data["TRENT_S_n"].mean(),
        "n_r_ecg_p_05": data["n_r_ecg_p_05"].median(),
        "ant_im": data["ant_im"].mean(),
    }

    for col, value in additional_columns.items():
        if col in data.columns:
            data[col].fillna(value, inplace=True)

    return data

# Columns to process
columns_to_copy = [
'AST_BLOOD', 'R_AB_1_n', 'NA_R_3_n', 'ZSN_A', 'NOT_NA_2_n', 'R_AB_3_n', 'S_AD_KBRIG',
       'MP_TP_POST', 'TRENT_S_n', 'np_09', 'ritm_ecg_p_06', 'NOT_NA_1_n', 'ANT_CA_S_n',
       'S_AD_ORIT', 'ROE', 'nr_07', 'D_AD_KBRIG', 'NOT_NA_3_n', 'NA_R_1_n', 'n_p_ecg_p_09',
       'lat_im', 'nr_01', 'NA_R_2_n', 'ant_im', 'nr_04', 'n_r_ecg_p_05', 'endocr_01', 'GB',
       'ritm_ecg_p_02', 'NOT_NA_KB', 'R_AB_2_n', 'ALT_BLOOD', 'n_r_ecg_p_06', 'np_10',
       'zab_leg_01', 'DLIT_AG', 'AGE'
]

# Ensure selected columns are a DataFrame
selected_columns = selected_data[columns_to_copy]

# Impute missing values
selected_data = impute_columns(selected_columns)

# Check for remaining missing values
print("Remaining missing values:\n", selected_data.isna().sum())
```

```python
def impute_missing_values(data):
    """
    Impute missing values for specified columns using mean, median, or mode.
    """
    data = data.copy()

    # Define imputation strategies for each column
    columns_to_impute = {
        "ritm_ecg_p_06": "mode",  # Use mode for categorical/binary data
        "D_AD_KBRIG": "median",  # Use median for numeric data with skew or outliers
        "n_p_ecg_p_09": "mode",  # Use mode for categorical/binary data
        "np_10": "mean"          # Use mean for continuous numeric data
    }

    # Apply imputation strategies
    for col, strategy in columns_to_impute.items():
        if strategy == "mean":
            data[col].fillna(data[col].mean(), inplace=True)
        elif strategy == "median":
            data[col].fillna(data[col].median(), inplace=True)
        elif strategy == "mode":
            data[col].fillna(data[col].mode()[0], inplace=True)

    return data

# Impute missing values for the specified columns
selected_data = impute_missing_values(selected_data)

# Check remaining missing values
print("Remaining missing values:\n", selected_data.isna().sum())
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

def impute_with_regression(data, target_col):
    """
    Impute missing values in the target column using linear regression with all other columns as predictors.
    """
    # Create a copy of the data to avoid modifying the original
    data = data.copy()

    # Convert all columns to numeric if possible
    data = data.apply(pd.to_numeric, errors='coerce')

    # Separate predictors (all columns except the target column)
    input_cols = [col for col in data.columns if col != target_col]

    # Separate rows with and without missing values in the target column
    known_data = data[data[target_col].notna()]
    missing_data = data[data[target_col].isna()]

    # If no missing values, return the original data
    if missing_data.empty:
        print(f"No missing values to impute for {target_col}.")
        return data

    # Prepare data for training and prediction
    X_train = known_data[input_cols]
    y_train = known_data[target_col]
    X_missing = missing_data[input_cols]

    # Train Linear Regression model
    regressor = LinearRegression()
    regressor.fit(X_train, y_train)

    # Predict missing values and fill them in the original data
    data.loc[data[target_col].isna(), target_col] = regressor.predict(X_missing)

    return data

# Target column to impute
target_col = "S_AD_KBRIG"

# Apply Linear Regression Imputation
selected_data = impute_with_regression(selected_data, target_col)

# Check remaining missing values
print("Remaining missing values:\n", selected_data.isna().sum())
```

# Plot and Evaluate Function

```python

# Define a function to train and evaluate models with a custom threshold
def train_and_evaluate_model(model, model_name, X_train, y_train, X_valid, y_valid, threshold=0.4):
    # Train the model
    model.fit(X_train, y_train)

    # If the model supports predict_proba, calculate probabilities for predictions
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_valid)[:, 1]
    else:
        # For SVM with linear kernel, use decision_function for probabilities
        y_pred_prob = model.decision_function(X_valid)

    # Apply custom threshold for classification
    y_pred = (y_pred_prob >= threshold).astype(int)

    # Evaluate the model
    print(f"--- {model_name} ---")
    print(f"Using Threshold: {threshold}")
    print("Accuracy Score on Validation Set:", accuracy_score(y_valid, y_pred))
    print("\nClassification Report on Validation Set:\n", classification_report(y_valid, y_pred))
    print("\nConfusion Matrix on Validation Set:\n", confusion_matrix(y_valid, y_pred))

    # Compute ROC curve and AUC
    fpr, tpr, thresholds = roc_curve(y_valid, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    # Plot ROC-AUC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random Guess (AUC = 0.50)')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title(f'{model_name} - Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.grid()
    plt.show()

# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42),
    "SVM (Linear Kernel)": SVC(kernel="linear", probability=True, random_state=42)
}

```

```python
def save_test_results(model, model_name, X_test, y_test, threshold=0.4, output_file="test_results.csv"):
    """
    Saves the test set predictions and probabilities to a CSV file and prints evaluation metrics.

    Parameters:
        model: Trained model to use for predictions.
        model_name (str): Name of the model.
        X_test: Features of the test set.
        y_test: True labels of the test set.
        threshold (float): Custom threshold for classification.
        output_file (str): Path to save the results CSV.
    """
    # If the model supports predict_proba, calculate probabilities for predictions
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    else:
        # For SVM with linear kernel, use decision_function for probabilities
        y_pred_prob = model.decision_function(X_test)

    # Apply custom threshold for classification
    y_pred = (y_pred_prob >= threshold).astype(int)

    # Evaluate the model
    print(f"--- {model_name} ---")
    print(f"Using Threshold: {threshold}")
    print("Accuracy Score on Test Set:", accuracy_score(y_test, y_pred))
    print("\nClassification Report on Test Set:\n", classification_report(y_test, y_pred))
    print("\nConfusion Matrix on Test Set:\n", confusion_matrix(y_test, y_pred))

    # Save results to a CSV file
    results_df = pd.DataFrame({
        "True Label": y_test,
        "Predicted Label": y_pred,
        "Predicted Probability": y_pred_prob
    })
    results_df.to_csv(output_file, index=False)
    print(f"\nTest results saved to: {output_file}")
```

```python

# Train and evaluate each model with the custom threshold
for model_name, model in models.items():
    train_and_evaluate_model(model, model_name, X_train_smote_admission, y_train_smote, X_test_admission, y_test, threshold=0.4)
```

## Comparison across different time periods

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

# Example: Time-point specific datasets
datasets = {
    "Time[0]": X_train_smote_admission,  # Features available at admission
    "Time[1]": X_train_after24,  # Features available after 24 hours
    "Time[2]": X_train_after48,  # Features available after 48 hours
    "Time[3]": X_train_after72   # Features available after 72 hours
}

# Target variable (same for all time points)
y = y_train_smote

# Store feature importance for each time point
feature_importance_dict = {}

for time_point, X in datasets.items():
    print(f"Training model for {time_point}...")
    model = RandomForestClassifier(random_state=42, n_estimators=100)
    model.fit(X, y)

    # Extract feature importance
    feature_importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": model.feature_importances_
    }).sort_values(by="Importance", ascending=False)

    # Store the importance
    feature_importance_dict[time_point] = feature_importance
    print(f"Top features for {time_point}:\n", feature_importance.head(10))
```

```python
# Prepare data for visualization
import matplotlib.pyplot as plt

# Combine feature importance into a single DataFrame
importance_data = pd.DataFrame({
    "Time[0]": feature_importance_dict["Time[0]"].set_index("Feature")["Importance"],
    "Time[1]": feature_importance_dict["Time[1]"].set_index("Feature")["Importance"],
    "Time[2]": feature_importance_dict["Time[2]"].set_index("Feature")["Importance"],
    "Time[3]": feature_importance_dict["Time[3]"].set_index("Feature")["Importance"]
}).fillna(0)

# Convert importance values to percentages
importance_data *= 100  # Scale to percentage

# Select top features based on average importance across all time points
top_features = importance_data.mean(axis=1).sort_values(ascending=False).head(10).index

# Dictionary to map feature codes to full forms
feature_full_form = {
    "ZSN_A": "History of Heart Failure",
    "AGE": "Age",
    "lat_im": "Side Heart Attack",
    "ROE": "Inflammation Marker",
    "S_AD_KBRIG": "Blood Pressure at Admission",
    "AST_BLOOD": "Liver Enzyme (AsAT)",
    "S_AD_ORIT": "Blood Pressure in ICU",
    "ALT_BLOOD": "Liver Enzyme (AlAT)",
    "ant_im": "Front Heart Attack",
    "DLIT_AG": "High Blood Pressure"
}

# Replace feature codes with full names in top_features
top_features_full = [feature_full_form[feature] for feature in top_features]

# Rename rows in importance_data to full forms for better readability
importance_data_full = importance_data.rename(index=feature_full_form)

# Plot feature importance trends
ax = importance_data_full.loc[top_features_full].T.plot(
    kind="bar",
    figsize=(14, 8),
    width=0.8,  # Increase bar width
    color=['skyblue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']
)

# Customize the plot
ax.set_title("Top Features Importance Over Time", fontsize=16)
ax.set_xlabel("Time Points", fontsize=14)
ax.set_ylabel("Feature Importance (%)", fontsize=14)  # Y-axis shows percentage
ax.set_xticks(range(4))
ax.set_xticklabels(["Time at Admission", "After 24 Hours", "After 48 Hours", "After 72 Hours"], fontsize=12)
plt.legend(title="Features", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Removed bar labels

plt.tight_layout()
plt.show()
```

This graph shows how important features change over time for predicting Heart Failure (ZSN) complications.​

At admission, features like heart failure and age are the most critical.​

By 48 and 72 hours, clinical factors like liver enzymes (AsAT, AIAT) and blood pressure in ICU become more important.​
